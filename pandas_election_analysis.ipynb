#First Question
Determine the winning party for every election at the state level. A party which
has the highest total votes is considered the winner here.


```python
#defining a function for winning party
def winning_party(year):                                  
  w=pd.DataFrame(df[df['Election Year']==year].groupby(['Party']).Votes.sum()).sort_values('Votes',ascending=False).head(1)
  return w

#input a year to get the answer 
winning_party('2005-Feb')
```

#Sixth question

Determine the stronghold party-seats. A stronghold seat is one that is won by the same party in all the elections.

```python

dt = df[df['Position']==1]                         #Determining the first positions
win = pd.pivot_table(dt,values='Party',index= ['AC name'],columns=['Election Year'],aggfunc={'Party':max})     #party who won in the corrosponding constituences       
dm = win[['2005-Feb','2005-Oct','2010','2015']]
same_party = win[dm.eq(dm.iloc[:, 0], axis=0).all(axis=1)]              #parties who won all the elections in same constituencies
same_party
```
